```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python

# Función f(x, y) que quieres aproximar (debes definirla)
def f(x, y):
    res=1+((1/2)*((-2)*(x**2)+(-2)*(y**2)))
    return res

```

```python
# Punto alrededor del cual realizarás la aproximación
a = 0.0
b = 0.0

# Rango de valores para x e y
x_range = np.linspace(a - 5, a + 5, 100)  # Intervalo pequeño alrededor de a
y_range = np.linspace(b - 5, b + 5, 100)  # Intervalo pequeño alrededor de b
X,Y=np.meshgrid(x_range,y_range)
Z=[f(x,y) for x,y in zip(X.ravel(),Y.ravel())]
Z = np.array(Z).reshape(X.shape)
print(Z.shape)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```
